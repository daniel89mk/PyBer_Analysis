```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# When we use the plt.plot() function to create a scatter plot, we need to add a lowercase "o" as a parameter 
# inside the parentheses. This switches the plot from a line chart to a scatter plot.
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(x_axis, y_axis)
```

```python
plt.scatter(y_axis, x_axis, color="red", label="Chicago")
plt.gca().invert_yaxis()
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.xlim(0,50)
```

```python
plt.scatter(x_axis, y_axis)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# To change the size of our marker on the graph, we have to use the s, or size parameter, for the marker. 
# The s parameter must be equal to a feature of our data. 
# We will make the size of each marker equal to the data floating-point decimal values of the y-axis by using s=y_axis.
# Some of the markers look very small 
# Let's make them bigger so that the chart is more legible and impactful
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis)
plt.show()
```

```python
# Add a percentage to one decimal place to each wedge of the pie.
# Increase the figure size to 8x8.
# Explode the two highest percentage months.
# Add a shadow.
# Add a start angle so that January is at the top.
# Reverse the order so that the month order is in a clockwise direction.
# Add new colors of your own choosing or use the colors from the previous pie chart.

colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    startangle=95,
    counterclock=False,
    shadow=True,
    autopct='%.1f%%')

plt.show()
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev)
```

```python
# We can make the error bars look more professional by adding a "cap" to the error bars that's equal to a scalar value. 
# To do this, we use the capsize= parameter.
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```
